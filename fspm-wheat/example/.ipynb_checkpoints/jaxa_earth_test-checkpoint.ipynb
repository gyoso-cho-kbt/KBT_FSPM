```python
# reference URL:
# [1] https://data.earth.jaxa.jp/en/
# [2] https://data.earth.jaxa.jp/api/python/jaxa.earth.html
from jaxa.earth import je
```

```python
data_out = je.ImageCollection("JAXA.EORC_ALOS.PRISM_AW3D30.v3.2_global")\
             .filter_date(["2021-01-01T00:00:00","2023-01-31T00:00:00"])\
             .filter_resolution(1)\
             .filter_bounds([135.20, 30.00, 145.28, 40.0])\
             .select("DSM")\
             .get_images()
```

```python
img = je.ImageProcess(data_out)\
        .show_images()
```

```python
collect_list = je.ImageCollectionList(ssl_verify=False)
col_id, col_band =  collect_list.filter_name(['SMC','daily'])
print(col_id)
print(col_band)
```

```python
data = je.ImageCollection(collection=col_id[2], ssl_verify=False)\
         .filter_date(["2023-08-30T00:00:00","2023-08-31T23:59:59"])\
         .filter_resolution()\
         .filter_bounds([135, 30, 140, 40])\
         .select(band='SMC')\
         .get_images()
```

```python
img = je.ImageProcess(data).show_images()
```
